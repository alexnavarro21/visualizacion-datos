```python
years = np.arange(2010, 2021)
countries = df["Country"].unique()
data = []

np.random.seed(42)
for country in countries:
    base = df[df["Country"] == country][["Economy (GDP per Capita)", "Family", "Health (Life Expectancy)"]].mean().values
    for i, year in enumerate(years):
        vals = base + np.random.normal(0, 0.5, 3) * base * (1 + 0.03 * i)
        data.append([country, year] + list(vals))

df_time = pd.DataFrame(data, columns=["Country", "Year", "Economy", "Family", "Health"])
df_time
```

```python
# VisualizacÃ³n NÂº6

df_time_region = df_time.merge(df[["Country", "Region"]], on="Country", how="left")

df_area = df_time_region.groupby(["Year", "Region"])[["Economy", "Family", "Health"]].mean().reset_index()

pivot = df_area.pivot(index="Year", columns="Region", values="Economy")

plt.figure(figsize=(12,7))
plt.stackplot(
    pivot.index,
    [pivot[region].fillna(0) for region in pivot.columns],
    labels=pivot.columns,
    alpha=0.85
)
plt.xlabel("AÃ±o")
plt.ylabel("Economy (GDP per Capita) promedio")
plt.title("Stacked Area Chart: EvoluciÃ³n de Economy por RegiÃ³n")
plt.legend(loc="upper left", bbox_to_anchor=(1.02, 1), ncol=1)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**QuÃ©:** El tipo de dataset son datos escalares (PIB promedio por regiÃ³n y aÃ±o).

**Por quÃ©:** Mostrar la evoluciÃ³n del PIB promedio entre las distintas regiones del mundo a lo largo del tiempo.

Tarea: Mostrar la evoluciÃ³n del PIB promedio por regiÃ³n.  
AcciÃ³n: Visualizar el cambio a lo largo de los aÃ±os.  
Objetivo: Apreciar cÃ³mo varÃ­a el PIB promedio en cada regiÃ³n y comparar tendencias.

**CÃ³mo:** Se usan las siguientes codificaciones visuales: Encode (Align, Color (Hue, Saturation)). La posiciÃ³n horizontal representa el aÃ±o, la posiciÃ³n vertical muestra el PIB promedio, y el color diferencia cada regiÃ³n en el grÃ¡fico de Ã¡reas apiladas.

```python
# VisualizaciÃ³n NÂº7
from scipy import stats

def gaussian_smooth(x, y, grid, sd):
    weights = np.transpose([stats.norm.pdf(grid, m, sd) for m in x])
    weights = weights / weights.sum(0)
    return (weights * y).sum(1)

# Prepara los datos reales de df_time por regiÃ³n
df_time_region = df_time.merge(df[["Country", "Region"]], on="Country", how="left")
df_area = df_time_region.groupby(["Year", "Region"])[["Economy"]].mean().reset_index()
pivot = df_area.pivot(index="Year", columns="Region", values="Economy")

x = pivot.index.values
grid = np.linspace(x.min()-4, x.max()+4, num=300)
y = [pivot[region].fillna(0).values for region in pivot.columns]
y_smoothed = [gaussian_smooth(x, y_, grid, 1) for y_ in y]

fig, ax = plt.subplots(figsize=(12, 7))
ax.stackplot(grid, y_smoothed, baseline="sym", labels=pivot.columns, alpha=0.85, cmap='Blues')
plt.xlabel("AÃ±o")
plt.ylabel("Economy (GDP per Capita) suavizado")
plt.title("Streamgraph: EvoluciÃ³n suavizada y centrada de Economy por RegiÃ³n")
plt.legend(loc="upper left", bbox_to_anchor=(1.02, 1), ncol=1)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**QuÃ©** El tipo de dataset son datos escalares.

**Por quÃ©:** Mostrar la evoluciÃ³n suavizada y centrada del PIB entre las distintas regiones del mundo a lo largo del tiempo.

Tarea: Mostrar la evoluciÃ³n suavizada del PIB por regiÃ³n.  
AcciÃ³n: Visualizar el cambio y las tendencias.  
Objetivo: Apreciar cÃ³mo varÃ­a el PIB suavizado en cada regiÃ³n y comparar patrones.

**CÃ³mo:** Se usan las siguientes codificaciones visuales: Encode (Align, Color (Hue, Saturation)). La posiciÃ³n horizontal representa el aÃ±o, la posiciÃ³n vertical muestra el PIB suavizado, y el color diferencia cada regiÃ³n en el streamgraph.

```python
# Visualizacion NÂº11

from pandas.plotting import parallel_coordinates
from sklearn.cluster import KMeans

# Selecciona las columnas numÃ©ricas relevantes
features = [
    "Economy (GDP per Capita)", "Family", "Health (Life Expectancy)",
    "Freedom", "Trust (Government Corruption)", "Generosity", "Dystopia Residual"
]
df_numeric = df.set_index("Country")[features]

# Normaliza los datos
df_scaled = (df_numeric - df_numeric.mean()) / df_numeric.std()

# Usa KMeans para encontrar 10 paÃ­ses lo mÃ¡s distintos posible
kmeans = KMeans(n_clusters=10, random_state=42)
df_scaled['cluster'] = kmeans.fit_predict(df_scaled)
selected_countries = df_scaled.groupby('cluster').apply(lambda x: x.sample(1)).index.get_level_values(1)
df_selected = df.loc[df['Country'].isin(selected_countries), ['Country'] + features]

# Parallel coordinates plot
plt.figure(figsize=(14,6))
parallel_coordinates(df_selected, 'Country', colormap=plt.get_cmap("tab10"))
plt.title("Parallel Coordinates Plot: Relaciones multivariadas en 10 paÃ­ses distintos")
plt.ylabel("Valor normalizado")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

**QuÃ©:** El tipo de dataset son datos escalares (factores normalizados de felicidad en 10 paÃ­ses distintos).

**Por quÃ©:** Mostrar las relaciones multivariadas y comparar los perfiles de factores de felicidad entre paÃ­ses seleccionados por su diversidad.

Tarea: Mostrar los factores normalizados en paÃ­ses diversos.  
AcciÃ³n: Visualizar las diferencias y similitudes.  
Objetivo: Apreciar cÃ³mo varÃ­an los factores de felicidad entre paÃ­ses muy distintos.
